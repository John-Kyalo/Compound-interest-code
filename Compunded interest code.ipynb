```python
# Import matplotlib.pyplot for plotting
import matplotlib.pyplot as plt

# Define a function to calculate the final balance after 30 years
def final_balance(savings, interest):
    balance = 0 # Initial balance
    for year in range(1, 31): # Loop over 30 years
        balance += savings * 12 # Add the savings for the year
        balance *= (1 + interest / 100) # Add the interest for the year
    return balance # Return the final balance

# Define the initial savings per month
savings = 100

# Define the interest rates
interest_rates = [1, 3, 5]

# Create an empty list to store the final balances
balances = []

# Loop over the interest rates
for interest in interest_rates:
    # Calculate the final balance for each interest rate
    balance = final_balance(savings, interest)
    # Append the balance to the list
    balances.append(balance)

# Print the final balances
print("The final balances after 30 years are:")
for i in range(len(interest_rates)):
    print(f"With {interest_rates[i]}% interest, the final balance is ${balances[i]:.2f}")

# Create an empty list to store the years
years = []

# Loop over 30 years
for year in range(1, 31):
    # Append the year to the list
    years.append(year)

# Create a figure and an axis for plotting
fig, ax = plt.subplots()

# Loop over the interest rates
for i in range(len(interest_rates)):
    # Create an empty list to store the balances for each year
    yearly_balances = []
    # Loop over the years
    for year in years:
        # Calculate the balance for each year
        yearly_balance = final_balance(savings, interest_rates[i]) / (1 + interest_rates[i] / 100) ** (30 - year)
        # Append the balance to the list
        yearly_balances.append(yearly_balance)
    # Plot the yearly balances for each interest rate
    ax.plot(years, yearly_balances, label=f"{interest_rates[i]}% interest")

# Set the title, labels, and legend for the plot
ax.set_title("Growth of savings over 30 years")
ax.set_xlabel("Year")
ax.set_ylabel("Balance ($)")
ax.legend()

# Save the plot as a PNG file
plt.savefig("savings_plot.png")
```
